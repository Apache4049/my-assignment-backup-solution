```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants?search_str=delhi')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_tittle = soup.find('div',class_="restnt-info cursor")
first_tittle
```

```python
first_tittle.text
```

```python
loc = soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
rating = soup.find('div',class_="restnt-rating rating-4")
rating
```

```python
rating.text
```

```python
img = soup.find('div',class_="img cursor")
img
```

```python
img.text
```

```python
sta = soup.find('span',class_="double-line-ellipsis")
sta.text
```
